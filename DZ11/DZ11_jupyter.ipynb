### Строим график

```python
x = list(range(-100, 100))
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from sympy import *

x1 = list(range(-100, 100))
a = -12
b = -10
c = 5
d = 10
e = -30
fx = [a*i**4*sin(cos(i)) + b*i**3 + c*i**2 + d*i + e for i in x1]

plt.axes(facecolor='white')  #заливка фона графика
plt.plot(x1,fx)
plt.show()
```
